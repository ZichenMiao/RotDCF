# Show Rotation Equivariance

```python
M1 = 1
M2 = 1
M3 = 1
Ntheta = 8

Nrot = 16
```

## Layer 1

```python
from Rot_DCF import Rot_DCF_Init

rot_dcf_init = Rot_DCF_Init(M2, M1, 5, Ntheta, 5, 3, bias=False)
```

### show rotate kernels

```python
import matplotlib.pyplot as plt

## with shape [Ntheta*chn_out, chn_in, k_s, k_s]
kernel = rot_dcf_init.rototion_weight()
kernel = kernel.view(Ntheta, M2, M1, 5, 5)
kernel = kernel[:, 0, 0, ...]
plt.figure(figsize=(10,5))
for i in range(Ntheta):
    plt.subplot(1,Ntheta,i+1)
    plt.imshow(kernel[i].detach().numpy())
    plt.axis('off')
plt.savefig('./rotate_filter.png')
```

### show rotation equivariance in output feature maps of first layer

```python
## test outputs of rotated images
## prepare data
import numpy as np
import torch
import PIL

import torchvision.transforms as transforms

idx_img = 3
data_path = '/home/zichen/Projects/RotDCF-Pytorch/data/MNIST/processed/training.pt'
per_rot_ang = 360/Nrot

## load data
data = torch.load(data_path)
images, labels = data

img_ori = images[idx_img]
img_ori_show = transforms.functional.to_pil_image(img_ori)


## create rotation images of angles [0, 22.5, ..., 337.5]
img_rot_show_list = []
img_rot_list = []

ang = 0
while ang < 360:
    img_rot_show = transforms.functional.rotate(img_ori_show, ang, resample=PIL.Image.BICUBIC)
    img_rot = transforms.functional.to_tensor(img_rot_show)
    img_rot_show_list.append(img_rot_show)
    img_rot_list.append(img_rot)
    ang += per_rot_ang

feat_map_1_list = []
for img_rot in img_rot_list:
    feat_map_1_list.append(rot_dcf_init(img_rot.unsqueeze(0).float()).view(Ntheta*M2, 28, 28).detach().numpy())

plt.figure(figsize=(40, 30))

## plt original image
for i in range(len(img_rot_show_list)):
    plt.subplot(Ntheta+1, Nrot, i+1)
    plt.imshow(img_rot_show_list[i], cmap='gray')
    plt.axis('off')
    if i == 0:
        plt.title('original image', fontsize=15)

# plt feat with specific \lambda(idx_feature_map) and \alpha(idx_rotation_group)
for alpha in range(Ntheta):
    for idx in range(len(feat_map_1_list)):
        plt.subplot(Ntheta+1, Nrot, (alpha+1)*Nrot+idx+1)
        ## fetch the features for a rot image
        feat = feat_map_1_list[idx]
        feat = feat.reshape([Ntheta, M2, 28, 28])
        plt.imshow(feat[alpha, 0], cmap='gray')
        plt.axis('off')
        if idx == 0:
            plt.title('alpha={}'.format(alpha+1), fontsize=15)
```

```python
from Rot_DCF import Rot_DCF, Rot_DCF_Init

rot_dcf = Rot_DCF(M3, M2, 5, Ntheta, 5, 3, bias=False)
```

### show rotate kernels

```python
import matplotlib.pyplot as plt

## with shape [Ntheta*chn_out, Ntheta*chn_in, k_s, k_s]
kernel = rot_dcf.rotation_weight()
kernel = kernel.view(Ntheta, M3, Ntheta, M2, 5, 5)
kernel = kernel[:, 0, :, 0, ...]

plt.figure(figsize=(8, 8))
cnt = 1
for i in range(Ntheta):
    for j in range(Ntheta):
        plt.subplot(Ntheta,Ntheta,cnt)
        # every row is rotated filters operated on input alpha channels
        plt.imshow(kernel[j, i].detach().numpy())
        
        plt.axis('off')
        cnt += 1
```
